```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("Animal protection and culling.csv")
```

```python
df.head()
```

```python
df.info()
```

- statistical Analysis

```python
df.describe()
```

- Adoption and Euthanasia Comparison for dogs

```python
plt.plot(df['year'], df['adopt a stray dog'], label='Adopt a Stray Dog', marker='o')
plt.plot(df['year'], df['euthanize dog'], label='Euthanize Dog', marker='o')
plt.legend() 
plt.title('Year vs Adoption and Euthanasia')
plt.xlabel('Year'); plt.ylabel('Count')
plt.show()
```

-  Adoption and Euthanasia Comparison for cats

```python
plt.plot(df['year'], df['adopt a stray cat'], label='Adopt a stray cat', marker='o')
plt.plot(df['year'], df['euthanize cat'], label='euthanize cat', marker='o')
plt.legend() 
plt.title('Year vs Adoption and Euthanasia')
plt.xlabel('Year'); plt.ylabel('Count')
plt.show()
```
